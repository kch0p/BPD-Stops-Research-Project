```python
ripa_df = ripa_df[ripa_df.LAT != -361]
ripa_df = ripa_df[ripa_df.LONG != -361]

ripa_df['LAT'].index.size
```

```python
from geopy.distance import geodesic
newport_ri = (41.49008, -71.312796)
cleveland_oh = (41.499498, -81.695391)
print(geodesic(newport_ri, cleveland_oh).miles)
```

```python
# %%timeit
# all_stops_small_copy = all_stops.head(8400)


# times_to_it = len(all_stops_small_copy.index)


i = 0
distance_list = []
uni_coords = (37.8719,122.2585)

for i in range(ripa_df.index.size):
    long_1 = ripa_df['LAT'][i]
    lat_1 = (ripa_df['LONG'][i])*-1
    stop_coords = (long_1,lat_1)
    distance = (geodesic(stop_coords, uni_coords).miles)
#     y = y + 1
    distance_list.append(distance)
#     if i >= 64420:
#         print('debug', (geodesic(stop_coords, uni_coords).miles),y, len(distance_list))

```

```python
len(distance_list)
```

```python
ripa_df['Distance from Cal'] = distance_list
ripa_df
```

```python
stop_point = [Point(ripa_df['LONG'][i], ripa_df['LAT'][i]) for i in np.arange(len(ripa_df))]
tt = [berkeley['geometry'].contains(stop_point[i]) for i in np.arange(len(stop_point))]
judge = [np.where(tt[i] == True)[0] for i in np.arange(len(tt))]
tract_column = [i[0] if i.size > 0  else -1 for i in judge]
ripa_df['tract'] = tract_column
berkeley['order'] = np.arange(len(berkeley))
ripa_df = ripa_df.merge(berkeley[['name10', 'order']], left_on = 'tract', right_on = 'order')
ripa_df = ripa_df.drop(columns=['tract','order'])
```

```python
ripa_df = ripa_df.rename(columns={"name10":"Census Tract"})
ripa_df.index.size
ripa_df
```

```python
filepath = Path('formatted_allstops.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
ripa_df.to_csv(filepath, na_rep='')
```
